```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.geeksforgeeks.org')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
for i in soup.find_all('a'):
    print(i.get('href'))
```

```python
for i in soup.find_all('a'):
    
    link = i.get('href')
    
    if('https' in link):
        print(link)
```

```python
links = []

for i in soup.find_all('a'):
    link = i.get('href')
    
    if('https' in link):
        links.append(link)
```

```python
links
```

```python
fd = open('links_test.txt','a')

for link in links:
    fd.write(link + '\n')
    
fd.close()
```
