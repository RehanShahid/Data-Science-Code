```python
import requests

res = requests.get('https://www.geeksforgeeks.org')
```

```python
res
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.
```

* The `prettify()` method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each tag and each string:

```python
soup.prettify
```

```python
with open('abc_temp.html','wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
def download(link, name):
    
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    with open(name,'wb') as file:
        file.write(soup.prettify('utf-8'))
```

```python
download('https://www.google.com','google.html')
```
